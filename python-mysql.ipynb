```python
df=pd.read_sql("""
select emp_id,emp_name,dept_name,mgr_name,proj_name
from employee e -- alias for table
left join 
department d
on e.dept_id=d.dept_id
left join 
manager m
on e.mgr_id=m.mgr_id
left join 
project p
on e.emp_id=p.team_member_id;
""",engine)
df
```

### to sql

```python
df.to_sql(name='joins',con=engine,index=False)
```
